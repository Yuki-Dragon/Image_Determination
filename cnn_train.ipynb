```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras.utils import np_utils
import numpy as np
import keras

image_name_list = ['lizard','dragon','seadragon']
num_list = len(image_name_list)
image_pxl = 50
num_used = 300
epoch =100
data_name = "dragon_series"
save_data_path = './' + data_name + '.npy'


def main():
    x_train, x_test, y_train, y_test = np.load(save_data_path, allow_pickle=True)
    # normalization by 256(RGB)
    x_train = x_train.astype("float")/256    
    x_test = x_test.astype("float")/256
    # one_hot vector for labels
    y_train = np_utils.to_categorical(y_train, num_list)
    y_test = np_utils.to_categorical(y_train, num_list)
    
    model = model_train(x_train, y_train,data_name,25)
    model_eval = (model,x_test,y_test)
    
    
def model_train(x_train,y_ind,data_name,epoch):
    model = Sequential()
    #convolute 3*3 pixels
    model.add(Conv2D(32,(3,3), padding='same', input_shape= x_train.shape[1:])) #shape of x_train =[ num of images/channels, pix, pix,3(=rgb)]
    # https://teratail.com/questions/94576 : good ref  http://aidiary.hatenablog.com/entry/20170216/1487252452:
    model.add(Activation('relu'))
    model.add(Conv2D(32,(3,3)))
    model.add(Activation('relu'))
    # half shrink
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))
    
    model.add(Conv2D(64,(3,3), padding='same'))
    model.add(Activation('relu'))
    model.add(Conv2D(64,(3,3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))    

    model.add(Flatten())
    model.add(Dense(512))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))
    model.add(Dense(3))
    # output
    model.add(Activation('softmax'))
    
    
    opt = keras.optimizers.rmsprop(lr=0.0001, decay=0.000001)
    
    model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
    
    model.fit(x_train,y_ind, batch_size =128, epochs=epoch)
    saved_cnn_name = './' + data_name + "_cnn.h5"
    model.save(saved_cnn_name)
    
    return model


def model_eval(model,x,y):
    scores = model.evaluate(x,y, verbose=1)
    print("Loss :",scores[0])
    print("Accuracy: ",scores[1])
    
    
    
if __name__ == "__main__":
    main()
    
    
```
