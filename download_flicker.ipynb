```python
from flickrapi import FlickrAPI
from urllib.request import urlretrieve
from pprint import pprint
import os, time, sys

key = '999f16c00a0db0bbd23c5e038aec8d74'
secret_key = '30483c7556d87ac2'

wait_time = 1

image_name_list = ['lizard','dragon','seadragon']

for i in range(len(image_name_list)):
    
    image_name = image_name_list[i] #sys.argv[1] # '0': filename 1~:input
    savedir = "./" + image_name


    flickr = FlickrAPI(key,secret_key, format="parsed-json")
    result = flickr.photos.search(
        text = image_name,
        per_page = 500,
        media = 'photos',
        sort = 'relevance',
        safe_search = 1,
        extras = 'url_q, licence'
        )

    photos = result['photos']
    
    if os.path.exists(savedir)!=1:
        directory = image_name
        parent_dir = "./"
        path = os.path.join(parent_dir, directory)  
        os.mkdir(path)  

    for ind, photo in enumerate(photos['photo']):
        _url_q = photo['url_q']
        filepath = savedir + '/' + photo['id'] + ".jpg"

        if os.path.exists(filepath): 
            continue

        urlretrieve(_url_q,filepath)
        time.sleep(wait_time)
        
```
